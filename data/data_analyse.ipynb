```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.head(5)
```

```python
train_df.info()
```

```python
def col_count(df, col_name, width=1.0, x_axis=[]):
    df_temp = df[col_name].value_counts()
    print(df_temp)
    fig = plt.figure(figsize=(8, 6))
    if len(x_axis) == 0:
        x_axis = df_temp.index
    y_axis = df_temp
    plt.bar(range(len(y_axis)), y_axis, width=width, color='rgb',tick_label=x_axis)
    plt.show()
```

```python
col_count(train_df, 'click')
```

```python
columns = train_df.columns.tolist()
columns.remove('id')
columns.remove('hour')
for column in columns:
    df_temp = train_df[column].value_counts()
    print(df_temp)
    print('\n\n')
```

```python
del train_df['id']
del train_df['hour']
train_df.to_csv('train_drop_id_hour.csv', index=0, header=0)  # 不保存列名，不保存行号
```
